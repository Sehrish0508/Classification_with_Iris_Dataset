## Choosing the best model for classification on Iris Dataset
- Logistic Regression
- Naive Bayes
- kNN Classifier
  
-Iris dataset can be extracted from sklearn directly 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
```

```python
# loading the dataset
df=load_iris(as_frame=True).frame
df.head(4)
```

```python
# Split the Data 
X=df.drop(columns="target")
y=df["target"]

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.25, random_state=42)
# Since there is no missing value so no need to handle it
```

## Exploratory Data Analysis - EDA

```python
# pie chart for target variable
classes_count=df["target"].value_counts()
print(classes_count)

# pie plot to visualize it properly
plt.pie(classes_count, labels=["Setosa", "Versicolor", "Virginica"], autopct="%1.1f%%")
plt.title("Categoried of Iris flower")
plt.show()

# bar plot to visualize it 
ax= sns.barplot(classes_count)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sns.histplot(data=df, x="sepal length (cm)", kde=True)
```

```python
# most famous plot for iris dataset 
g = sns.pairplot(df, hue="target")

g._legend.set_title("Species")
plt.show()
```

## Train and Evaluate the models 
- kNN classifier
- Naive Bayes
- Logistic Regression

```python
# Training and evaluation with logistic regression
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
lr_model=LogisticRegression()
lr_model.fit(X_train, y_train)

y_pred =lr_model.predict(X_test)

# Evaluation
print("Logistic Regression Classifier Model")
print("Accuracy: ", accuracy_score(y_test, y_pred) )
print("CM:", confusion_matrix(y_test, y_pred) )
```

```python
knn_model= KNeighborsClassifier(n_neighbors=1)   # decided k by hit n trial here 
knn_model.fit(X_train, y_train)

y_pred =knn_model.predict(X_test)

# Evaluation
print("kNN Classifier Model")
print("Accuracy: ", accuracy_score(y_test, y_pred) )
print("CM:", confusion_matrix(y_test, y_pred) )
```

```python
# Naive Bayes

nb_model= GaussianNB()
nb_model.fit(X_train, y_train)

y_pred =nb_model.predict(X_test)

# Evaluation

print("Naive Bayes Classifier Model")
print("Accuracy: ", accuracy_score(y_test, y_pred) )
print("CM:", confusion_matrix(y_test, y_pred) )
```

## All 3 models performed exceptionally well because iris is a very clean noise free dataset
